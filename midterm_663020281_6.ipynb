```python
print("ก่อนกรอง:", df_all.shape[0])
print("หลังกรอง:", df_all_new.shape[0])
print("หายไป:", df_all.shape[0] - df_all_new.shape[0])
```

```python
cols = ['tele_station_oldcode']

missing_rows = df_all.merge(
    df_all_new[cols],
    on=cols,
    how='left',
    indicator=True
).query('_merge == "left_only"').drop(columns='_merge')
```

```python
missing_rows
```

```python
df_all_u = df_all.drop_duplicates(subset=cols)
missing_u = missing_rows.drop_duplicates(subset=cols)
```

```python
total_per_station = (
    df_all_u
    .groupby('tele_station_oldcode')
    .size()
    .reset_index(name='total_rows')
)
```

```python
missing_per_station = (
    missing_u
    .groupby('tele_station_oldcode')
    .size()
    .reset_index(name='missing_rows')
)
```

```python
summary = total_per_station.merge(
    missing_per_station,
    on='tele_station_oldcode',
    how='left'
).fillna(0)

summary['missing_ratio'] = summary['missing_rows'] / summary['total_rows']
summary['missing_percent'] = summary['missing_ratio'] * 100

summary.sort_values('missing_percent', ascending=False)
```

##5.เติมข้อมูลที่หายไป

```python
rh
```

```python
import pandas as pd

# To avoid SettingWithCopyWarning, explicitly work on a copy if df_all_new was created as a slice
df_all_new = df_all_new.copy()

# 1. Prepare df_all_new: Convert 'measure_datetime' to datetime and extract 'month' and 'year'
df_all_new['measure_datetime'] = pd.to_datetime(df_all_new['measure_datetime'])
df_all_new['month'] = df_all_new['measure_datetime'].dt.month
df_all_new['year'] = df_all_new['measure_datetime'].dt.year

# 2. Prepare rh for merging
# Rename 'Node' to 'tele_station_oldcode' to match df_all_new
rh_for_merge = rh.rename(columns={'Node': 'tele_station_oldcode'})
# Keep only relevant columns for merging and the rain value
rh_for_merge = rh_for_merge[['tele_station_oldcode', 'month', 'year', 'rain']]

# 3. Merge rh_for_merge into df_all_new to get the monthly rain values
df_all_new = pd.merge(
    df_all_new,
    rh_for_merge,
    on=['tele_station_oldcode', 'month', 'year'],
    how='left' # Use left merge to keep all rows from df_all_new
)

# Rename the merged 'rain' column to be more descriptive
df_all_new.rename(columns={'rain': 'monthly_rain_from_rh'}, inplace=True)


# 4. Fill missing 'rainfall_value' in df_all_new using 'monthly_rain_from_rh'
# Only fill if 'rainfall_value' is NaN
df_all_new['rainfall_value'] = df_all_new['rainfall_value'].fillna(df_all_new['monthly_rain_from_rh'])

# 5. Clean up: Drop the temporary columns if no longer needed
df_all_new.drop(columns=['monthly_rain_from_rh', 'month', 'year'], inplace=True)

print("เติมข้อมูลที่หายไปในคอลัมน์ 'rainfall_value' ของ df_all_new ด้วยค่าปริมาณน้ำฝนรายเดือนจาก rh แล้วครับ")
print("ตรวจสอบข้อมูลที่หายไปหลังจากเติม:")
display(df_all_new['rainfall_value'].isnull().sum())
print("แถวที่มีค่า 'rainfall_value' เป็น NaN หลังจากการเติม:")
display(df_all_new[df_all_new['rainfall_value'].isnull()])

display(df_all_new.head())
```

##6.อภิปราย เหตุผลที่่ใช้ median แทน mean

**ความทนทานต่อค่าผิดปกติ (Outliers):**

**Mean (ค่าเฉลี่ย):** ได้รับผลกระทบอย่างมากจากค่าผิดปกติ หรือข้อมูลที่มีค่าสูง/ต่ำกว่าปกติมากๆ เช่น ถ้าปริมาณน้ำฝนส่วนใหญ่เป็น 0-10 มม. แต่มีอยู่ 1 วันที่ฝนตก 1000 มม. ค่าเฉลี่ยจะถูกดึงให้สูงขึ้นอย่างผิดปกติ ซึ่งอาจทำให้ค่าเฉลี่ยไม่สะท้อนปริมาณน้ำฝน 'ทั่วไป' ได้ดีนัก

**Median (มัธยฐาน):** เป็นค่ากลางที่เกิดจากการเรียงลำดับข้อมูลจากน้อยไปมาก และเลือกค่าที่อยู่ตรงกลางพอดีทำให้ไม่ได้รับผลกระทบจากค่าผิดปกติมากเท่าค่าเฉลี่ย จึงเป็นตัวแทนของข้อมูลส่วนใหญ่ได้ดีกว่าเมื่อข้อมูลมี Outliers

**สรุป**
ข้อมูลปริมาณน้ำฝนมักจะมีการกระจายตัวที่ไม่สมมาตรคือมีค่า 0 หรือค่าน้อยๆ จำนวนมาก และมีค่าสูงๆ เพียงไม่กี่ค่านกรณีเช่นนี้ Medianจะเป็นค่ากลางที่เหมาะสมกว่า Mean เพราะ Mean จะถูกดึงไปทางด้านที่มีค่าน้อยหรือมากกว่าตามลักษณะการเบ้ของข้อมูล




##7.คำนวณปริมาณฝนรายเดือน

```python
import pandas as pd

# Ensure 'measure_datetime' is in datetime format
df_all_new['measure_datetime'] = pd.to_datetime(df_all_new['measure_datetime'])

# Extract year and month
df_all_new['year'] = df_all_new['measure_datetime'].dt.year
df_all_new['month'] = df_all_new['measure_datetime'].dt.month

# Calculate monthly rainfall by grouping and summing
monthly_rainfall = df_all_new.groupby(['tele_station_oldcode', 'year', 'month'])['rainfall_value'].sum().reset_index()

# Rename the summed column for clarity
monthly_rainfall.rename(columns={'rainfall_value': 'monthly_total_rainfall'}, inplace=True)

print("คำนวณปริมาณน้ำฝนรายเดือนสำหรับแต่ละสถานีแล้วครับ:")
display(monthly_rainfall.head())

# Optionally, display some basic info about the new DataFrame
print("\nข้อมูลของ DataFrame monthly_rainfall:")
monthly_rainfall.info()
```

```python
import pandas as pd

# Define the list of specific stations to include
specific_stations = ['ACRU', 'BBHN', 'BBUA', 'YOM012']

# Filter df_all_new to include only these specific stations
df_specific_stations = df_all_new[df_all_new['tele_station_oldcode'].isin(specific_stations)].copy()

# Ensure 'measure_datetime' is in datetime format (it should already be from previous steps, but re-ensure for safety)
df_specific_stations['measure_datetime'] = pd.to_datetime(df_specific_stations['measure_datetime'])

# Extract year and month
df_specific_stations['year'] = df_specific_stations['measure_datetime'].dt.year
df_specific_stations['month'] = df_specific_stations['measure_datetime'].dt.month

# Calculate monthly rainfall by grouping and summing for specific stations
monthly_rainfall_specific = df_specific_stations.groupby(['tele_station_oldcode', 'year', 'month'])['rainfall_value'].sum().reset_index()

# Rename the summed column for clarity
monthly_rainfall_specific.rename(columns={'rainfall_value': 'monthly_total_rainfall'}, inplace=True)

print("คำนวณปริมาณน้ำฝนรายเดือนสำหรับ 4 สถานีที่เลือกแล้วครับ:")
display(monthly_rainfall_specific.head(10))

print("\nข้อมูลของ DataFrame monthly_rainfall_specific:")
monthly_rainfall_specific.info()
```
