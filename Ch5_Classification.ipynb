```python
from sklearn.datasets import load_digits

# Load the MNIST dataset
digits = load_digits()
x,y = digits.data, digits.target
print('shape of data:', (x.shape))
print('shap of target :', (y.shape))
print('number of classes :',(len(digits.target_names)))
print('first 5 target label :',(y[:5]))
```

# Model Construction
- import
- define
- train
- evaluate (train , test)

###data train

```python
x_train = x[:1700]
y_train = y[:1700]
x_test = x[1700:]
y_test = y[1700:]
```

###import

```python
from sklearn.tree import DecisionTreeClassifier
```

###Define

```python
tree0= DecisionTreeClassifier(random_state=6)
```

###Trainning

```python
tree0.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(20,10))
plot_tree(tree0, filled=True, feature_names=digits.feature_names, class_names=[str(x) for x in digits.target_names], rounded=True)
plt.title("Decision Tree Visualization")
plt.show()
```

###Evaluate

```python
y_predict = tree0.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

##Quiz 11

tune hyper-parameters so that the accuracy is better than default (st random_state = 6)

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
tree3 = DecisionTreeClassifier(
    random_state=6,
    max_depth= None,
    criterion='entropy',      # ลองใช้ entropy แทน gini เพื่อการวัดข้อมูลที่ละเอียดขึ้น
    min_samples_split=2,      # ป้องกันไม่ให้แยกกิ่งกับข้อมูลที่น้อยเกินไป
    min_samples_leaf=2,       # มั่นใจว่าแต่ละใบมีข้อมูลสนับสนุนอย่างน้อย 2 ตัว
    class_weight='balanced'   # ถ้าข้อมูลแต่ละกลุ่มมีจำนวนไม่เท่ากัน ตัวนี้ช่วยได้มาก
)

tree3.fit(x_train, y_train)
y_predict3 = tree3.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, y_predict3)}")
```
